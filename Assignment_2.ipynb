```python
# Question 1

import numpy as np

arr = np.random.rand(10,1)
print(arr)
print("Mean: ",np.mean(arr))
```

```python
# Question 2

import numpy as np

arr = np.random.rand(20,1)
print(arr)
print("Variance: ",np.var(arr))
print("Standard Deviation: ",np.std(arr))
```

```python
# Question 3

import numpy as np

A = np.random.rand(10,20)
B = np.random.rand(20,25)

arrMul = np.matmul(A,B)
print(arrMul)
print("Shape of new matrix: ",np.shape(arrMul))
print("Sum of all elements: ",np.sum(arrMul))
```

```python
# Question 4

import numpy as np

Arr =np.random.rand(10,1)

print("Random Array: ",Arr)

print('*'*40)

def fxn(x):
    return (1 / (1 + np.exp(-x)))

NewArr = np.apply_along_axis(fxn,0,Arr)

print("New Array :",NewArr)
```
